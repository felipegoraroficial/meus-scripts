```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('data/gas_prices.csv')
```

```python
plt.title('Valores do Combustivel em US$')

plt.plot(gas['Year'], gas['Australia'], 'r.-', label= 'Australia')
plt.plot(gas['Year'], gas['Italy'], 'y.-', label= 'Italy')
plt.plot(gas['Year'], gas['USA'], 'b.-', label= 'USA')
plt.plot(gas['Year'], gas['Japan'], 'g.-', label= 'Japan')

plt.xticks(gas['Year'][::2])

plt.xlabel('Year')
plt.ylabel('US$')

plt.legend()
```

```python
plt.title('Valores do Combustivel em US$')

list_county = ['Australia','USA','Italy', 'Mexico','France','Germany']

for country in gas:
    if country in list_county:
        plt.plot(gas['Year'], gas[country], label= [country], marker='.')

plt.xticks(gas['Year'][::2])

plt.xlabel('Year')
plt.ylabel('US$')

plt.legend()
plt.savefig('data/country.png')
```
